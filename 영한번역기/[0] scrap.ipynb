# 네이버 사전에서 영어-한글 예문 말뭉치 긁어오기

```python
from urllib.request import urlopen, HTTPError
from bs4 import BeautifulSoup
import re
dictionary = open('dictionary.txt', 'r',encoding='UTF8')
word = dictionary.readline()
#비상! 중간에 중단해서 다시 시작해야함. 다음은 고치는 코드
#while word != "ejaculation\n":
#    word = dictionary.readline()

corpus = open('corpus_after.txt','w', encoding='UTF8')
continuetime = 0
maxcontinue = 5
while word :
    print(word[:-1])
    try :
        url = "https://endic.naver.com/search_example.nhn?sLn=kr&query=" + word[:-1] + "&examType=example"
        html = urlopen(url)
    except :
        if continuetime < maxcontinue :
            continuetime += 1
            print("--retry--")
            continue
        else :
            print("--loading failed--")
    else :
        if html == None :
            print("URL was not found")
        else :
            bsObj = BeautifulSoup(html, "html.parser")
            #print(bsObj, '\n', '='*30)
            utb = bsObj.findAll("li", {"class":re.compile("utb")})
            for i in range(len(utb)) :
                try :
                    translated = utb[i].find("div",{"class":"fnt_k10"}).find("a",{"class":"N=a:xmp.detail"}).get_text()
                except :
                    pass
                else :
                    if translated != None :
                        corpus.write(utb[i].find("input").attrs["value"] + '\n')
                        corpus.write(translated + '\n')
    word = dictionary.readline()
dictionary.close()
corpus.close()
print('--end--')
```
